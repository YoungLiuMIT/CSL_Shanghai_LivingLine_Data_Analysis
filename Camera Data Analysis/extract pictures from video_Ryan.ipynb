```python
import cv2
import os
```

```python
videos_src_path = r'C:\Users\RYAN\Google Drive\MAS\Tongji D&I\LivingLine Data\Camera Data\CSS Pavilion Data\test data\001_for people detection algorithms\set 3'
videos_save_path = r'C:\Users\RYAN\Google Drive\MAS\Tongji D&I\LivingLine Data\Camera Data\CSS Pavilion Data\test data\001_for people detection algorithms\set 3\extracted images'
os.chdir(videos_src_path)
vc = cv2.VideoCapture('ML Test Video.mp4') #读入视频文件
```

```python
if (vc.isOpened()== False): 
    print('Error opening video stream or file')
    exit()
```

```python
timeF = 15    #视频帧计数间隔频率
frame_get = 0
timeS = 0
os.chdir(videos_save_path)
while frame_get<3600:    #循环读取视频帧 #设置结束的帧数
    rval, frame = vc.read()
    if rval == True:
        frame_get += 1
        if frame_get % timeF == 0:    #每隔timeF帧进行存储操作
            timeS = int(frame_get/30)    #int is floor
            image_file_name = 'ML Test Video_{}_{}.png'.format(str(timeS), str(frame_get % 30))
            cv2.imwrite(image_file_name, frame) #存储为图像
            #cv2.waitKey(0)
            print('{} is saved'.format(image_file_name))
vc.release()
```
