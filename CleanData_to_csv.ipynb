```python
import urllib
import json
import datetime
import re
import csv

def read_data():
    f = open("pri_2018_07_05_02","r", encoding='utf-8', errors='ignore')
    data = f.readlines()
    f.close()
    return data

def read_json():
    f = open("ApIndex.json","r", encoding='utf-8', errors='ignore')
    indexList = f.read()
    f.close()
    return indexList

def time_formatted(time):
    timeCorrection=8*60*60
    dateTime=datetime.datetime.utcfromtimestamp(float(time)/1000 + timeCorrection).strftime('%Y/%m/%d %H:%M:%S')
    return dateTime

def get_line(x, indexList):
    lineData = []
    x = x.split(',')
    lineData.append(x[4])
    time_receive = time_formatted(x[4])
    lineData.append(time_receive)
    lineData.append(x[7])
    time_send = time_formatted(x[7])
    lineData.append(time_send)
    lineData.append(x[5])
    lineData.append(x[9])
    wifiList = x[25]
    count = wifiList.count('ssid')
    lineData.append(count)
    lineData.append(x[34])
    for a in indexList:
        if(a['MAC_Address'] == x[5]):
            lineData.append(a['index'])
            break
    return lineData

def write_data(data):
    with open('output.csv', 'a', errors='ignore', newline='') as f:
            f_csv = csv.writer(f)
            f_csv.writerows(data)
    
if __name__ == '__main__':
    csvData = []
    csvData.append(['ts_receive', 'ts_receive_time', 'ts_send', 'ts_send_time', 'ap_mac', 'device_mac', 'count', 'ap_type', 'ap_index'])
    data = read_data()
    indexList = read_json()
    indexList = json.loads(indexList)
    data_formatted = '[' + re.sub('}\s*{',  '},{', data[0]) + ']'
    data_formatted = re.sub("'tongji",  "{'tongji", data_formatted)
    data_formatted = re.sub("上海市,'",  "上海市,'}", data_formatted)
    data_json = json.loads(data_formatted)
    for line in data_json: 
        dataList = line['data']
        for x in dataList: 
            lineData = get_line(x, indexList)
            csvData.append(lineData)
    write_data(csvData)
    print ('finished')
```
