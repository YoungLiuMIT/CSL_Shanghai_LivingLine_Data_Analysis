```python
import urllib
import json
import datetime

def read_data():
    f = open("2018_07_15_04","r", encoding='utf-8', errors='ignore')
    data = f.readlines()
    f.close()
    return data
    
def get_lineResult(line):
    JSON_data = json.loads(line)
    wifi = JSON_data['wifidata']
    test_result = {}
    test_result["ap_mac"] = wifi['apmac']
    test_result["count"] = wifi['count']
    test_result["ts_send"] = wifi['tssend']
    dateTime=datetime.datetime.utcfromtimestamp(float(wifi['tssend'])/1000).strftime('%Y/%m/%d %H:%M:%S')
    test_result["time_send"] = dateTime
    wifiList = wifi['wifitalist']
    mac_list=[]
    for x in wifiList:
        temp0 = {}
        temp0["mac"] = x['mac']
        temp0["rssi"] = x['rssi']
        mac_list.append(temp0)
    test_result["mac_list"] = mac_list
    return test_result

def get_result(data):
    result = list()
    for line in data:
        line_result = get_lineResult(line)
        result.append(line_result)
    return result

def writeFile(result):
    JSON_result = json.dumps(result, indent=4)
    f = open("data_prettified.txt", "w")
    f.write(JSON_result)
    f.close
    print("finished")

if __name__ == '__main__':
    data = read_data()
    result = get_result(data)
    writeFile(result)
```
